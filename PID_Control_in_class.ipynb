```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
def qin(t):
    return 0.2
```

```python
def dh(t, v):
    h, eint = v
    e = h - hsp
    qmv = qmv0 + kc*(e + eint/taui)
    return [(qin(t) - cv*h**0.5 +qmv)/A,
            e]
```

```python
tend = 3
A = 1
cv = 1
kc = -5
hsp = 1.5
taui = 0.01

eint0 = 0.
qmv0 = 0.2
h0 = ((qin(0)+qmv0)/cv)**2
hsp=h0*2
print(f'h0={h0}    hsp={hsp}')
```

```python
sol = solve_ivp(dh, (0,tend), np.array([h0, eint0]), dense_output=True, method='Radau').sol
t_plot = np.linspace(0,tend, 200)
h, eint  = sol(t_plot)
e = h - hsp
qmv = qmv0 + kc*(e + eint/taui)
fig=make_subplots(rows=1,cols=2)
fig.add_scatter(x=t_plot ,y=h, row=1,col=1,name='height')
fig.add_scatter(x=t_plot, y=qmv, row=1,col=2,name='qm')
```
