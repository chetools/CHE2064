```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.optimize import root
```

```python
def radau(N):
    x=np.zeros(N)
    w=np.zeros(N)
    x[0]=-1
    w[0]=2/N**2
    x[1:]=((np.polynomial.Legendre.basis(N-1)+np.polynomial.Legendre.basis(N))//np.polynomial.Polynomial([1,1]).convert(kind=np.polynomial.Legendre)).roots()
    w[1:]=(1-x[1:])/(N*np.polynomial.Legendre.basis(N-1)(x[1:]))**2
    return x,w
```

```python
def lobatto(N):
    x=np.zeros(N)
    w=np.zeros(N)
    x[0]=-1
    x[-1]=1
    w[0]=w[-1]=2/N/(N-1)
    x[1:-1]=(np.polynomial.Legendre.basis(N-1).deriv()).roots()
    w[1:-1]=2/N/(N-1)/(np.polynomial.Legendre.basis(N-1)(x[1:-1]))**2
    return x,w
```

```python
def eqs(v,x):
    Y=np.polynomial.Chebyshev(v,domain=(0,5))
    return np.concatenate([np.array([Y(0)-2, Y.deriv()(0)-1.]), (Y.deriv(2)+4*Y)(x)])
```

```python
x,w=np.polynomial.legendre.leggauss(10)
x=(x+1)*5/2
```

```python
v0=np.ones(12)
```

```python
eqs(v0,x)
```

```python
sol=root(lambda v: eqs(v,x),v0)
```

```python
sol
```

```python
xplot=np.linspace(0,5,100)
fig=make_subplots()
Y=np.polynomial.Chebyshev(sol.x,domain=(0,5))
fig.add_scatter(x=xplot,y=Y(xplot),mode='markers')
fig.add_scatter(x=xplot,y=(np.sin(2*xplot)+4*np.cos(2*xplot))/2,mode='lines')
```
